```python
a_p_d['TOTAL_EMPLOYED_YEAR'] = a_p_d.apply(lambda x: abs(round(x['DAYS_EMPLOYED'] / 365,2)), axis=1)
a_p_d['TOTAL_BIRTH_AGE_YEAR'] = a_p_d.apply(lambda x: abs(round(x['DAYS_BIRTH'] / 365,2)), axis=1)
a_p_d['TOTAL_REGISTRATION_YEAR'] = a_p_d.apply(lambda x: abs(round(x['DAYS_REGISTRATION'] / 365,2)), axis=1)
a_p_d['TOTAL_ID_PUBLISH_YEAR'] = a_p_d.apply(lambda x: abs(round(x['DAYS_ID_PUBLISH'] / 365,2)), axis=1)
```
